```python
import matplotlib.pyplot as plt
import pandas as pd

def plot_multiple_time_series(file_paths, title='HWMId observed Anomaly with respect to years', xlabel='Years', ylabel='Anomaly'):

    plt.figure(figsize=(15, 8))

    for i, file_path in enumerate(file_paths):
        # Read the dataset
        df = pd.read_csv(file_path+'.csv')

        # Convert year to datetime by setting to the first day of the year
        df['time'] = pd.to_datetime(df['time'].astype(int), format='%Y')


        mean = df['HWMId'][23 : 52].mean()
        df['anomaly'] = df['HWMId'] - mean
        # Set date column as index
        #df.set_index('time', inplace=True)

        # Plot the time series data
        plt.plot(df['time'][0 : 82],df['anomaly'][0 : 82], linestyle ='-', label=f'{file_path}')

    plt.title(title)
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    plt.legend()
    plt.grid(True)
    plt.savefig('anomaly.png')
    plt.show()
```

```python
name = ['Ad-Damazin_HWMId', 'Kassala_HWMId', 'Nyala_HWMId', 'Singa_HWMId', 'Al_Qadarif_HWMId',
        'Khartoum_HWMId', 'Omdurman_HWMId','El_ObeidHWMId', 'Kosti_HWMId', 'Ouadi_Halfa_HWMId']
# Plot the datasets
plot_multiple_time_series(name)
```
