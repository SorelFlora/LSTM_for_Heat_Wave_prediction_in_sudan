```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
def Heat_Wave_Frequency(name):
  for i, data in enumerate(name):

    df = pd.read_csv(data)
    df['time'] = pd.to_datetime(df['time'].astype(int), format='%Y')
    #df.set_index('time', inplace=True)

    plt.figure(figsize=(10, 6))
    plt.plot(df['time'][45:74],df['HWF'][45:74], color ='r', linestyle = '-')
    plt.xlabel('Year')
    plt.ylabel('Heat wave frequency')
    plt.title('Variation of HWF with respect to years')
    plt.grid(True)
    plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd

def plot_multiple_time_series(file_paths, title='HWF observed anomaly with respect to years', xlabel='Year', ylabel='Anomaly'):

    plt.figure(figsize=(15, 10))

    for i, file_path in enumerate(file_paths):
        # Read the dataset
        df = pd.read_csv(file_path+'.csv')

        # Convert year to datetime by setting to the first day of the year
        df['time'] = pd.to_datetime(df['time'].astype(int), format='%Y')

        mean = df['HWF'][23 : 52].mean()
        df['anomaly'] = df['HWF'] - mean
        # Set date column as index
        #df.set_index('time', inplace=True)

        # Plot the time series data
        plt.plot(df['time'][0 : 82],df['anomaly'][0 : 82], linestyle ='-', label=f'{file_path}')

    plt.title(title)
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    plt.legend()
    plt.grid(True)
    plt.savefig('HWF anomaly.png')
    plt.show()
```

```python
name = ['Ad-Damazin_HWF', 'El_Obeid_HWF', 'Khartoum_HWF', 'Nyala_HWF', 'Ouadi_Halfa_HWF',
        'Al_Qadarif_HWF', 'Kassala_HWF', 'Kosti_HWF', 'Omdurman_HWF', 'Singa_HWF']
# Plot the datasets
plot_multiple_time_series(name)
```
